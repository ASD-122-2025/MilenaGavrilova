```python
Відповіді на контрольні запитання:
1. Що таке pip і conda та чим вони відрізняються між собою?

pip (Pip Installs Packages) — це менеджер пакетів для Python, який використовується для встановлення та керування пакетами, написаними на Python. Він зосереджений на пакетах Python, доступних у Python Package Index (PyPI).
conda — це менеджер пакетів і середовищ з відкритим кодом, який може встановлювати пакети з будь-якої мови (включаючи Python, R, C++ тощо). Він також дозволяє створювати ізольовані середовища, що особливо корисно для управління залежностями в проектах, які використовують різні версії пакетів.
основні відмінності:

Сфера застосування: pip призначений виключно для пакетів Python, тоді як conda може керувати пакетами з різних мов.
Управління залежностями: conda краще справляється з управлінням залежностями, особливо з не-Python залежностями. Він також може встановлювати двійкові файли, що може бути корисним для деяких пакетів.
Середовища: conda має вбудовану підтримку для створення ізольованих середовищ, тоді як для pip потрібно використовувати інструменти, такі як virtualenv або venv.
2. Як виглядатиме виконання пунктів 1–3 індивідуального завдання за допомогою Virtualenv?

Створення віртуального середовища:
Відкрийте термінал або командний рядок.
Перейдіть до каталогу, де ви хочете створити віртуальне середовище.
Виконайте команду: virtualenv myenv (де "myenv" — назва вашого середовища).
Активація віртуального середовища:
На Windows: myenv\Scripts\activate
На macOS і Linux: source myenv/bin/activate
Встановлення необхідних пакетів за допомогою pip:
Після активації середовища використовуйте pip для встановлення необхідних пакетів. Наприклад: pip install pandas numpy matplotlib
Встановлення пакетів за допомогою Conda (якщо потрібно):
Якщо вам потрібно встановити пакети які не можливо встановити за допомогою pip, можна скористатись conda.
Наприклад: conda install tensorflow
Деактивація віртуального середовища:
Після завершення роботи з середовищем виконайте команду: deactivate
3. Сфотографувати себе на телефон та імпортувати своє зображення у форматі .png на початок робочого зошита.

Для цього вам потрібно сфотографувати себе на телефон, а потім перенести зображення на комп'ютер. Це можна зробити за допомогою USB-кабелю, хмарних сервісів або електронної пошти.
Після цього ви можете вставити зображення у форматі .png у свій робочий зошит, використовуючи функцію вставки зображення у вашому текстовому редакторі або середовищі розробки.
4. Написати процедуру на Python сортування масиву методом бульбашки.

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


5. Як виконати компіляцію вхідного документа Jupyter Notebook з його середовища, не використовуючи інтерфейс командного рядка?

Ви можете використовувати розширення nbconvert для Jupyter Notebook.
Відкрийте Jupyter Notebook.
Перейдіть до "File" -> "Download as" і виберіть потрібний формат (наприклад, HTML, PDF, LaTeX).
6. Як виконати компіляцію вхідного документа у Quarto?

Quarto має вбудовану підтримку для компіляції документів.
Відкрийте документ Quarto у редакторі (наприклад, VS Code з розширенням Quarto).
Натисніть кнопку "Render" у редакторі або виконайте команду quarto render your_document.qmd у терміналі.
Також можливо використовувати графічний інтерфейс Quarto Preview.
```
