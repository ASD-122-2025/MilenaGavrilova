## Контрольні питання

-----

### 1. Чим відрізняється структура бінарне дерево від бінарного дерева пошуку?

* **Бінарне дерево (Binary Tree):** Кожен вузол може мати не більше двох дочірніх вузлів (лівий та правий). Ніяких спеціальних правил щодо порядку розміщення значень у вузлах не існує.

* **Бінарне дерево пошуку (Binary Search Tree - BST):** Це особливий вид бінарного дерева, який задовольняє такі властивості для кожного вузла:
    * Значення у лівому піддереві вузла менші або рівні значенню у самому вузлі.
    * Значення у правому піддереві вузла більші за значення у самому вузлі.
    * Обидва ліве та праве піддерева також є бінарними деревами пошуку.

Основна відмінність полягає у **порядку розміщення елементів**, що робить бінарні дерева пошуку ефективними для операцій пошуку, вставки та видалення елементів.

-----

### 2. Чим відрізняється структура бінарне дерево від бінарної купи?

* **Бінарне дерево (Binary Tree):** Як описано вище, обмежується лише кількістю дочірніх вузлів.

* **Бінарна купа (Binary Heap):** Це спеціалізоване деревоподібне структура даних, що задовольняє властивість купи:
    * **Для min-купи:** Значення кожного вузла більше або рівне значенню його батьківського вузла (корінь має найменше значення).
    * **Для max-купи:** Значення кожного вузла менше або рівне значенню його батьківського вузла (корінь має найбільше значення).
    * Купа також є **майже повним бінарним деревом**, тобто всі рівні, крім, можливо, останнього, повністю заповнені, а вузли останнього рівня заповнюються зліва направо.

Основні відмінності:
    * **Властивість впорядкування:** Бінарна купа має специфічну властивість впорядкування між батьківськими та дочірніми вузлами, тоді як бінарне дерево не має таких обмежень.
    * **Форма:** Бінарна купа є майже повним бінарним деревом, що забезпечує ефективне зберігання у масиві. Бінарне дерево може мати будь-яку форму.
    * **Призначення:** Бінарні купи часто використовуються для реалізації черг з пріоритетами та алгоритмів сортування (наприклад, Heap Sort), тоді як бінарні дерева є більш загальною структурою даних.

### 3. Які існують типи дерев? Опишіть їхні основні характеристики та переваги.

Існує багато типів дерев, ось деякі з основних:

* **Бінарне дерево (Binary Tree):**
    * **Характеристики:** Кожен вузол має не більше двох дочірніх вузлів (лівий та правий).
    * **Переваги:** Проста структура, основа для багатьох інших складніших структур даних.

* **Бінарне дерево пошуку (Binary Search Tree - BST):**
    * **Характеристики:** Значення в лівому піддереві менші або рівні, в правому - більші.
    * **Переваги:** Ефективні операції пошуку, вставки та видалення в середньому випадку ($O(\log n)$).

* **Збалансовані дерева пошуку (Balanced BSTs) - AVL-дерева, червоно-чорні дерева:**
    * **Характеристики:** Автоматично підтримують певну збалансованість висоти піддерев, щоб запобігти виродженню в лінійну структуру.
    * **Переваги:** Гарантована логарифмічна складність ($O(\log n)$) для операцій пошуку, вставки та видалення в найгіршому випадку.

* **B-дерева та B+-дерева:**
    * **Характеристики:** Самозбалансовані дерева, оптимізовані для дискових операцій. Кожен вузол може мати велику кількість дочірніх вузлів.
    * **Переваги:** Ефективні для зберігання та пошуку великих обсягів даних на жорстких дисках (використовуються в базах даних та файлових системах).

* **Префіксні дерева (Trie або Prefix Tree):**
    * **Характеристики:** Деревоподібна структура, де кожен вузол представляє префікс рядка. Шлях від кореня до вузла утворює рядок.
    * **Переваги:** Ефективний пошук слів з певним префіксом, автодоповнення, перевірка правопису.

* **Купа (Heap):**
    * **Характеристики:** Майже повне бінарне дерево, що задовольняє властивість купи (min-купа або max-купа).
    * **Переваги:** Ефективна реалізація черг з пріоритетами, швидке знаходження мінімального або максимального елемента ($O(1)$), основа для сортування Heap Sort ($O(n \log n)$).

### 4. Наведіть приклади задач, які ефективно вирішуються за допомогою дерев.

* **Пошук, вставка та видалення даних:** Бінарні дерева пошуку та збалансовані BST є ефективними для реалізації множин, словників та інших структур даних, що вимагають швидких операцій з даними.
* **Автодоповнення та пошук за префіксом:** Префіксні дерева (Trie) ідеально підходять для реалізації функцій автодоповнення, пошуку слів за початковими літерами.
* **Керування пріоритетами:** Купи використовуються для реалізації черг з пріоритетами, де елементи обробляються в залежності від їхнього пріоритету.
* **Сортування:** Алгоритм Heap Sort використовує структуру купи для ефективного сортування масивів.
* **Представлення ієрархічних даних:** Дерева природно відображають ієрархічні структури, такі як файлові системи, організаційні структури, XML/JSON-документи.
* **Маршрутизація в комп'ютерних мережах:** Деревоподібні структури можуть використовуватися для організації маршрутизаційних таблиць.
* **Стиснення даних:** Дерева Хаффмана використовуються в алгоритмах стиснення даних.
* **Парсинг виразів:** Дерева розбору (parse trees) використовуються компіляторами для аналізу синтаксису програм.

### 5. Як організована купа? Опишіть алгоритми додавання та вилучення елементів з купи.

**Організація купи:**

Купа зазвичай реалізується як **майже повне бінарне дерево**, що також може бути ефективно представлене за допомогою **масиву**. Для вузла з індексом `i`:
* Лівий дочірній елемент знаходиться за індексом `2*i + 1`.
* Правий дочірній елемент знаходиться за індексом `2*i + 2`.
* Батьківський елемент знаходиться за індексом `(i - 1) // 2`.

Властивість купи (min-купа або max-купа) підтримується шляхом порівняння значень батьківських та дочірніх вузлів.

**Алгоритм додавання елемента (Heapify Up):**

1.  Додайте новий елемент в кінець купи (в наступну доступну позицію в масиві).
2.  Порівняйте доданий елемент з його батьківським елементом.
3.  Якщо властивість купи порушується (наприклад, у min-купі, якщо новий елемент менший за батьківський), поміняйте їх місцями.
4.  Продовжуйте порівнювати новий елемент з його новим батьківським елементом і міняти місцями, доки не буде відновлено властивість купи або не буде досягнуто кореня.

**Алгоритм вилучення максимального елемента (для max-купи) / мінімального елемента (для min-купи) (Heapify Down):**

1.  Вилучіть кореневий елемент (максимальний або мінімальний).
2.  Перемістіть останній елемент купи на місце кореня.
3.  Порівняйте новий кореневий елемент з його дочірніми елементами.
4.  Якщо властивість купи порушується (наприклад, у max-купі, якщо кореневий елемент менший за один з дочірніх), поміняйте його місцями з більшим (для max-купи) або меншим (для min-купи) дочірнім елементом. Якщо обидва дочірні елементи порушують властивість, оберіть більший (або менший) з них для обміну.
5.  Продовжуйте порівнювати новий елемент з його новими дочірніми елементами і міняти місцями, доки не буде відновлено властивість купи або не буде досягнуто листка.

### 6. Які задачі можна ефективно вирішити за допомогою купи? Наведіть приклади.

* **Черги з пріоритетами (Priority Queues):** Купи є природною та ефективною реалізацією черг з пріоритетами, де елементи з вищим пріоритетом обробляються першими. Приклади: керування завданнями в операційній системі, обслуговування клієнтів за рівнем важливості.
* **Алгоритм сортування Heap Sort:** Ефективний алгоритм сортування з часовою складністю $O(n \log n)$ в найгіршому випадку.
* **Знаходження k-го найбільшого/найменшого елемента:** Купу можна використовувати для ефективного знаходження k-го найбільшого або найменшого елемента в масиві за час $O(n \log k)$ або $O(k \log n)$.
* **Алгоритми на графах:** Купи використовуються в деяких алгоритмах на графах, таких як алгоритм Дейкстри для знаходження найкоротших шляхів та алгоритм Прима для побудови мінімального кістякового дерева.
* **Злиття k відсортованих списків:** Купу можна використовувати для ефективного злиття кількох відсортованих списків в один відсортований список.
* **Онлайн-алгоритми:** Купи корисні в онлайн-алгоритмах, де дані надходять поступово і потрібно підтримувати інформацію про мінімальні/максимальні елементи.

### 7. Як геш-функція використовується для зберігання та пошуку даних в геш-таблиці?

**Геш-функція (Hash Function)** — це функція, яка приймає вхідні дані (ключ) будь-якого розміру та перетворює їх у вихідні дані фіксованого розміру (геш-значення або геш-код). У контексті геш-таблиць геш-функція використовується для визначення **індексу (позиції)** в масиві (що становить основу геш-таблиці), де має зберігатися або звідки має бути знайдений елемент з даним ключем.

**Процес зберігання (вставки) даних:**

1.  Для заданого ключа обчислюється його геш-значення за допомогою геш-функції.
2.  Це геш-значення зазвичай приводиться до діапазону індексів масиву геш-таблиці за допомогою операції взяття за модулем (modulo).
3.  Елемент (ключ-значення) зберігається у комірці масиву за обчисленим індексом.

**Процес пошуку даних:**

1.  Для заданого ключа обчислюється його геш-значення за допомогою тієї самої геш-функції.
2.  Геш-значення знову приводиться до індексу масиву.
3.  Перевіряється комірка масиву за цим індексом. Якщо там знаходиться шуканий ключ (або елемент, пов'язаний з цим ключем), пошук завершується успішно.

**Важливість геш-функції:**

Якість геш-функції є критично важливою для ефективності геш-таблиці. Ідеальна геш-функція має такі властивості:

* **Рівномірний розподіл:** Геш-значення повинні бути рівномірно розподілені по всьому діапазону індексів масиву, щоб мінімізувати **колізії** (ситуації, коли різні ключі відображаються на один і той самий індекс).
* **Швидкість обчислення:** Геш-функція повинна бути швидкою для обчислення, оскільки вона викликається для кожної операції вставки та пошуку.

Незважаючи на зусилля щодо розробки хороших геш-функцій, колізії є неминучими, особливо коли кількість ключів перевищує розмір геш-таблиці. Тому в геш-таблицях використовуються різні **методи розв'язання колізій** (наприклад, ланцюжкове гешування, відкрита адресація).

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Створення дерева згідно з прикладом
root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(8)
root.left.left = TreeNode(1)
root.left.right = TreeNode(4)
root.right.left = TreeNode(6)
root.right.right = TreeNode(9)
```

```python
def delete_subtree(root, target_value):
    if root is None:
        return None

    if root.value == target_value:
        return None  # Видаляємо піддерево, встановивши посилання на None

    root.left = delete_subtree(root.left, target_value)
    root.right = delete_subtree(root.right, target_value)

    return root

# Приклад використання (видалення гілки, що починається з вузла зі значенням 3)
print("Дерево до видалення:")
# Тут потрібна функція для друку дерева (наприклад, обхід)
def print_tree_inorder(root):
    if root:
        print_tree_inorder(root.left)
        print(root.value, end=" ")
        print_tree_inorder(root.right)

print_tree_inorder(root)
print()

root = delete_subtree(root, 3)

print("Дерево після видалення гілки з вузлом 3:")
print_tree_inorder(root)
print()
```


− Оцінити асимптотичну складність (в середньому і в найгіршому випадку) процедур search, insert і delete роботи з деревом.
Для звичайного бінарного дерева (не обов'язково збалансованого):

Search (пошук):
Середній випадок: O(logn) - якщо дерево більш-менш збалансоване.
Найгірший випадок: O(n) - якщо дерево вироджується в лінійний список.
Insert (вставка):
Середній випадок: O(logn) - якщо місце вставки знаходиться на глибині, пропорційній логарифму кількості вузлів.
Найгірший випадок: O(n) - якщо дерево є лінійним, нова вставка може вимагати проходу через всі n вузлів.
Delete (видалення):
Середній випадок: O(logn) - пошук вузла для видалення та подальші операції (пошук наступника/попередника) займають час, пропорційний висоті дерева.
Найгірший випадок: O(n) - у виродженому дереві пошук вузла може зайняти O(n), а подальші операції також можуть бути дорожчими.
Для збалансованих бінарних дерев пошуку (наприклад, AVL, червоно-чорні):

Усі три операції (search, insert, delete) гарантовано мають асимптотичну складність O(logn) у середньому та найгіршому випадках завдяки механізмам балансування, які підтримують висоту дерева в логарифмічній залежності від кількості вузлів.



### − Написати процедуру генерації купи з будь-якого рандомного масива.
### − Додати до нього елемент, який дорівнює вашому порядковому номеру у списку групи.
### − Вилучити максимальний елемент з купи.
### − Оцінити асимптотичну складність (в середньому і в найгіршому випадку) процедур search, insert і delete роботи з купою.







```python
def heapify(arr):
    n = len(arr)
    # Побудова max-купи починаючи з останнього нелистового вузла
    for i in range(n // 2 - 1, -1, -1):
        _heapify_down(arr, n, i)

def _heapify_down(arr, n, i):
    largest = i  # Ініціалізуємо найбільший як корінь
    left = 2 * i + 1  # лівий = 2*i + 1
    right = 2 * i + 2  # правий = 2*i + 2

    # Перевіряємо, чи лівий дочірній елемент більший за корінь
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Перевіряємо, чи правий дочірній елемент більший за найбільший
    if right < n and arr[right] > arr[largest]:
        largest = right

    # Якщо найбільший не корінь
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # міняємо місцями
        # Рекурсивно обробляємо піддерево
        _heapify_down(arr, n, largest)

# Приклад використання
random_array = [3, 1, 4, 1, 5, 9, 2, 6]
heapify(random_array)
print(f"Купа з рандомного масиву: {random_array}")
```

```python
def heap_insert(heap, item):
    heap.append(item)
    _heapify_up(heap, len(heap) - 1)

def _heapify_up(heap, i):
    parent = (i - 1) // 2
    if parent >= 0 and heap[i] > heap[parent]:
        heap[i], heap[parent] = heap[parent], heap[i]
        _heapify_up(heap, parent)

# Додаємо елемент
your_number = 7
heap_insert(random_array, your_number)
print(f"Купа після додавання елемента {your_number}: {random_array}")
```

```python
def heap_extract_max(heap):
    if not heap:
        return None
    if len(heap) == 1:
        return heap.pop()

    max_element = heap[0]
    heap[0] = heap.pop()
    _heapify_down(heap, len(heap), 0)
    return max_element

# Вилучаємо максимальний елемент
max_val = heap_extract_max(random_array)
print(f"Максимальний елемент: {max_val}")
print(f"Купа після вилучення максимуму: {random_array}")
```

− Оцінити асимптотичну складність (в середньому і в найгіршому випадку) процедур search, insert і delete роботи з купою.
Search (пошук):
Середній випадок: O(n) - у купі немає властивості впорядкування для ефективного пошуку конкретного елемента, крім максимального/мінімального. У найгіршому випадку потрібно переглянути всі елементи.
Найгірший випадок: O(n).
Insert (вставка):
Середній випадок: O(logn) - додавання елемента в кінець та подальше просіювання вгору (heapify_up) займає час, пропорційний висоті купи.
Найгірший випадок: O(logn).
Delete (видалення):
Видалення максимального/мінімального (кореня):
Середній випадок: O(logn) - заміна кореня останнім елементом та подальше просіювання вниз (heapify_down) займає час, пропорційний висоті купи.
Найгірший випадок: O(logn).
Видалення довільного елемента: Вимагає пошуку елемента (O(n)), а потім відновлення властивості купи (O(logn)), тому загальна складність O(n).

До основних методів розв'язання колізій належать:

Роздільне ланцюгування (Separate Chaining): Кожна комірка геш-таблиці містить покажчик на зв'язаний список (або іншу структуру даних), куди зберігаються всі елементи, що мають однаковий геш-індекс.
Відкрита адресація (Open Addressing): Усі елементи зберігаються безпосередньо в геш-таблиці. При виникненні колізії шукається наступна вільна комірка за певним правилом (пробій). До методів відкритої адресації належать:
Лінійне пробіювання (Linear Probing): Послідовний пошук наступної вільної комірки.
Квадратичне пробіювання (Quadratic Probing): Пошук вільної комірки зі збільшенням кроку за квадратичною функцією.
Подвійне гешування (Double Hashing): Використання другої геш-функції для визначення кроку пробіювання.

2. Реалізуйте геш-таблицю з ланцюжковим гешуванням.

```python
class ChainedHashTable:
    def __init__(self, capacity):
        self.capacity = capacity
        self.table = [[] for _ in range(capacity)]
        self.size = 0

    def _hash(self, key):
        return hash(key) % self.capacity

    def insert(self, key, value):
        index = self._hash(key)
        for item in self.table[index]:
            if item[0] == key:
                item[1] = value  # Оновлюємо значення, якщо ключ вже існує
                return
        self.table[index].append([key, value])
        self.size += 1

    def search(self, key):
        index = self._hash(key)
        for item in self.table[index]:
            if item[0] == key:
                return item[1]
        return None

    def delete(self, key):
        index = self._hash(key)
        for i, item in enumerate(self.table[index]):
            if item[0] == key:
                self.table[index].pop(i)
                self.size -= 1
                return True
        return False

    def get_size(self):
        return self.size

# Приклад використання
ht = ChainedHashTable(10)
ht.insert(5, "value1")
ht.insert("hello", "value2")
ht.insert((1, 2), "value3")

print(f"Розмір геш-таблиці: {ht.get_size()}")
print(f"Пошук ключа 5: {ht.search(5)}")
print(f"Пошук ключа 'hello': {ht.search('hello')}")
print(f"Пошук ключа (1, 2): {ht.search((1, 2))}")
print(f"Пошук неіснуючого ключа: {ht.search(10)}")

ht.delete(5)
print(f"Розмір геш-таблиці після видалення: {ht.get_size()}")
print(f"Пошук ключа 5 після видалення: {ht.search(5)}")
```

3. Проведіть тестування геш-таблиці з різними типами даних (цілі числа, рядки, списки, словники, об’єкти):

```python
import time

def test_hash_table(ht, data):
    insert_times = []
    search_times = []
    delete_times = []

    for key, value in data.items():
        start_time = time.time()
        ht.insert(key, value)
        insert_times.append(time.time() - start_time)

    for key in data.keys():
        start_time = time.time()
        ht.search(key)
        search_times.append(time.time() - start_time)

    for key in list(data.keys()): # Щоб можна було ітеруватися та видаляти
        start_time = time.time()
        ht.delete(key)
        delete_times.append(time.time() - start_time)

    return {
        "insert_time": sum(insert_times) / len(insert_times) if insert_times else 0,
        "search_time": sum(search_times) / len(search_times) if search_times else 0,
        "delete_time": sum(delete_times) / len(delete_times) if delete_times else 0,
    }

# Тестування з різними типами даних
test_data = {
    1: "int_value",
    "hello": "str_value",
    (1, 2): "tuple_value",  # Змінено на кортеж, який є хешованим
    str({"a": 1}): "dict_value",  # Використовуємо рядкове представлення словника
    id(object()): "obj_value" # Використовуємо id об'єкта як хешований ключ
}

ht = ChainedHashTable(20)
results = test_hash_table(ht, test_data)

print("\nРезультати тестування:")
for op, avg_time in results.items():
    print(f"Середній час виконання {op}: {avg_time:.6f} секунд")

# Порівняння результатів:
print("\nПорівняння результатів:")
print("В цілому, час виконання основних операцій в геш-таблиці з ланцюжковим гешуванням має бути відносно сталим для різних хешованих типів даних (числа, рядки, кортежі).")
print("Для нехешованих типів даних (таких як словники та об'єкти, якщо використовувати їх безпосередньо), виникає помилка.")
print("При використанні рядкового представлення словника або id об'єкта як ключа, операції можливі, але ефективність може залежати від розподілу геш-значень цих представлень.")
print("Ефективність геш-таблиці сильно залежить від якості геш-функції та коефіцієнта заповнення (кількість елементів / розмір таблиці).")
```
