## Контрольні запитання

---

### 1. Надати визначення складності задачі із символом Ω.

Символ **Ω (Омега)** використовується в асимптотичному аналізі для опису **нижньої межі** складності алгоритму. Кажуть, що функція <span class="math-inline">T\(n\)</span> має складність <span class="math-inline">\\Omega\(f\(n\)\)</span>, якщо існують такі константи <span class="math-inline">c \> 0</span> та <span class="math-inline">n\_0 \\ge 0</span>, що для всіх <span class="math-inline">n \\ge n\_0</span> виконується нерівність:

<span class="math-block">T\(n\) \\ge c \\cdot f\(n\)</span>

Це означає, що час виконання алгоритму, починаючи з певного розміру вхідних даних (<span class="math-inline">n\_0</span>), завжди буде більшим або рівним деякій константі, помноженій на функцію <span class="math-inline">f\(n\)</span>. Іншими словами, <span class="math-inline">\\Omega</span>-нотація визначає найкращий можливий сценарій або мінімальну кількість операцій, які алгоритм повинен виконати.

### 2. Функція часової складності має вигляд: <span class="math-inline">F\(N\) \= N^3 \+ 7N^2 − 14N</span>. Як записати асимптотичну складність у O-нотації?

Для визначення асимптотичної складності в **O-нотації** ми враховуємо лише член найвищого порядку, оскільки він домінує при великих значеннях <span class="math-inline">N</span>. У функції <span class="math-inline">F\(N\) \= N^3 \+ 7N^2 − 14N</span> найвищим порядком є <span class="math-inline">N^3</span>. Тому асимптотична складність у O-нотації буде:

<span class="math-block">O\(N^3\)</span>

---

### 3. Функція часової складності має вигляд: <span class="math-inline">F\(N\) \= 1\.01N \+ N^\{10\}</span>. Як записати асимптотичну складність у O-нотації?

У функції <span class="math-inline">F\(N\) \= 1\.01N \+ N^\{10\}</span> член <span class="math-inline">N^\{10\}</span> зростає значно швидше, ніж <span class="math-inline">1\.01N</span> при великих значеннях <span class="math-inline">N</span>. Отже, асимптотична складність у O-нотації буде визначатися цим членом:

<span class="math-block">O\(N^\{10\}\)</span>

---

### 4. Функція часової складності має вигляд: <span class="math-inline">F\(N\) \= N^\{1\.3\} \+ 10 \\log\_2 N</span>. Як записати асимптотичну складність у O-нотації?

Порівнюючи члени <span class="math-inline">N^\{1\.3\}</span> та <span class="math-inline">10 \\log\_2 N</span>, поліноміальна функція <span class="math-inline">N^\{1\.3\}</span> зростає значно швидше за логарифмічну функцію <span class="math-inline">\\log\_2 N</span> при великих значеннях <span class="math-inline">N</span>. Тому асимптотична складність у O-нотації буде:

<span class="math-block">O\(N^\{1\.3\}\)</span>

---

### 5. У чому полягає ідея розпаралелювання обчислень і для чого вона використовується? Які з алгоритмів, наведених у цій лабораторній роботі, дозволяють можливість розпаралелювання?

**Ідея розпаралелювання обчислень** полягає у поділі складної обчислювальної задачі на кілька менших підзадач, які можуть виконуватися **одночасно** на кількох обчислювальних ресурсах (наприклад, ядрах процесора, графічних процесорах або кількох комп'ютерах).

**Використовується** розпаралелювання для значного **прискорення** виконання програм, особливо при обробці великих обсягів даних, у наукових симуляціях, машинному навчанні та інших ресурсомістких завданнях.

Щодо алгоритмів, наведених у цій лабораторній роботі, можливість розпаралелювання залежить від конкретних алгоритмів. Зазвичай, алгоритми, які мають незалежні цикли або операції над різними частинами даних, добре піддаються розпаралелюванню. Наприклад, якщо в лабораторній роботі розглядаються алгоритми сортування (наприклад, Merge Sort), матричні операції або певні види пошуку, вони можуть мати потенціал для розпаралелювання

### 6. Які існують способи підвищення обчислювальної швидкості алгоритмів? Який з них є найефективнішим?

Існує кілька способів підвищення обчислювальної швидкості алгоритмів:

* **Використання ефективних структур даних:** Правильний вибір структури даних може значно зменшити час виконання операцій (наприклад, пошуку, вставки, видалення).
* **Оптимізація коду:** Видалення зайвих обчислень, використання швидших операцій, розгортання циклів тощо.
* **Розпаралелювання обчислень:** Виконання кількох частин алгоритму одночасно.
* **Використання кешу:** Оптимізація доступу до даних для кращого використання кеш-пам'яті процесора.
* **Алгоритмічні покращення:** Розробка принципово нових, більш ефективних алгоритмів для розв'язання задачі.

Важко назвати один "найефективніший" спосіб, оскільки це залежить від конкретної задачі та обмежень. Однак, **розпаралелювання обчислень** часто є одним з найбільш потужних способів досягти значного прискорення, особливо для задач, які мають великий обсяг незалежних обчислень і виконуються на системах з багатьма обчислювальними ядрами або процесорами. Проте, ефективність розпаралелювання залежить від природи алгоритму та можливості ефективно розділити задачу. Алгоритмічні покращення, хоча і можуть бути складними, часто призводять до найбільш значних виграшів у швидкості для великих вхідних даних.
